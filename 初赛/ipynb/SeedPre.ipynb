```python
def getDict():
    a_label = []
    b_label = []
    c_label = []

    with open('train_a.txt') as f:
        isFirst = True

        while True:
            line = f.readline()

            if isFirst:
                isFirst = False
                continue            

            if not line:
                break

            l = line.split('\t')

            a_label.append(l[5])

            b_label.append(l[6])

            c_label.append(l[7].replace('\n',''))


    a_label = list(set(a_label))
    b_label = list(set(b_label))
    c_label = list(set(c_label))

    a_label.sort()
    b_label.sort()
    c_label.sort()
    
    a_dict = {}
    b_dict = {}
    c_dict = {}
    count = 1

    for i in a_label:
        a_dict[i] = str(count)
        count+=1

    for i in b_label:
        b_dict[i] = str(count)
        count+=1

    for i in c_label:
        c_dict[i] = str(count)
        count+=1


    print(a_dict)
    print(b_dict)
    print(c_dict)
    return (a_dict,b_dict,c_dict)
    
```

```python
def joinStrWithTab(l):
    str = l[0]
    for i in range(1,len(l)):
        str = str +'\t' +l[i]
    return str

def makeContin():
    
    target = open('train_a_c.txt','w')
    #target = open('valid_a_c.txt','w')
    
    with open('train_a.txt') as f:
   # with open('valid_a.txt') as f:
        isFirst = True

        while True:
            line = f.readline()

            if isFirst:
                isFirst = False
                continue            

            if not line:
                break

            #line.replace('\n','')
            l = line.split('\t')
                        
            l[5] = a_dict[l[5]]
            l[6] = b_dict[l[6]]
                    
            temp = l[7].replace('\n','');
            l[7] = l[7].replace(temp, c_dict[temp])
                                  
            target.write(joinStrWithTab(l))
               
    target.close()

makeContin()
```
